# What is a stack ?

Stack is a liner data structure that stores the items in LIFO manner.

so What are the operations we can perform on stack ?

# Basic operation of stack.

Let's see the below image to understand this.

image.png

Let's assume, we want to place item 1, 2, 3 in stack data structure. so,

* First we will take a empty stack.
* Let's put **1** in our stack. we call it **push** operation.
* Let's put **2** in our stack. see here 2 is placed on the top of 1.
* next we will put **3** in our stack.

ok. we are done with putting items in our stack. But what if we want to remove items from it.

we have to remove **3** first, then **2** & at last **1**.

Let's summarize our basic operations of stack -

* **push** - Add an element to the top of a stack.

* **pop** - Remove an element from the top of a stack.

in programming, we can also do following operations-

* **peek** - Get the value of the top element without removing it.

* **isEmpty** - Check if the stack is empty.

* **isFull** - Check if the stack is full.

# Implementation of Stack

There are many ways to implement stack. Let's see using inbuilt list.

## Using List


### Create an empty stack

```python
def create_stack():
  stack = []
  return stack

stack = create_stack()
print("Stack is created successfully!", stack)
```

**Notice** there is no element in our stack currently. so let's put some item in it.

### push an item to stack

```python
def push(stack, item):
  stack.append(item)
  print("Item %s pushed to stack sucessfully!" %(item))
```

```python
push(stack, 1)
push(stack, 2)
push(stack, 3)

print("\n Stack contains : ", stack)
```

see, we have inserted item 1, 2, 3 successfully to stack.

Now Let's pop an item.

## pop an item

before removing an item, we can check if our stack is empty.

```python
def pop(stack):
  if len(stack) == 0:
    return "Stack is Empty"
  else:
    return stack.pop()
```

```python
popped_item = pop(stack)

print("Popped Item : ", popped_item)
print("\n Stack after popping an item : ", stack)
```

Notice, the 3 (last item) was popped & there is no 3 in our stack. ok. But Let's suppose someone just wants the value of the last item, without removing it. To do that, we can peek an item.

## peek item 

we can get value of topmost element without removing it.

```python
def peek(stack):
  if stack:
    return stack[-1]    # get the last element of stack
```

```python
peek_item = peek(stack)
print("Item at the top of stack : ", peek_item)
print("Stack after peek an item : ", stack)
```

**see**, we had [1, 2] in our stack & the top element was 2, so we got value **2** & our stack remains the same.

# Time Complexity

So here all the mentioned operations takes constant time, hence time complexity is **O(1)**
